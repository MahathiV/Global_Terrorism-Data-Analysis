```python
#Number of terrorists that participated in the attack.

for i in range(len(df['nperps'])):
    if ((df['nperps'][i] == -9) or (df['nperps'][i] == -99)):
        df['nperps'][i] = "Unknown"
```

```python
df['country_txt'].nunique()
```

```python
# Group by with count

df_g = df.groupby('country_txt').count()
df_g.head()
```

```python
# reset the dataframe

df_g_rst = df_g.reset_index()
df_g_rst.head()
```

```python
# sorted the datframe in reverse order to get the highest number of attacked nations

df_g_sorted = df_g_rst.sort_values(by="eventid",ascending=False,ignore_index=True)
df_g_sorted.head()
```

```python
# first 25 countries with highest number of attacks

t_25 = df_g_sorted.head(25)
t_25
```

```python
# highest in top 25 countries

t_25['eventid'].max()
```

```python
# lowest in top 25 countries

t_25['eventid'].min()
```

```python
# list of top 25 countries 

top_25_countries = list(t_25['country_txt'])
print(top_25_countries)
```
